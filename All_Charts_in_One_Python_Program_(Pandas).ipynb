```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
#----
#---Simple dataset
#---
data={
    "Month":["Jan","Feb","Mar","Apr","May","Jun","Jul","Aug","Sep","Oct","Nov","Dec"],
    "sales":[2500,2500,3500,4500,2500,3600,4800,7800,8000,9800,4800,4900],
    "profits":[500,600,800,700,900,850,650,489,650,650,450,650]
}
df=pd.DataFrame(data)
df.head()
#--Line Charts
df.plot(x="Month",y="sales",kind="line",marker=0)
plt.title("Monthly sales Charts")
plt.ylabel("sales")
plt.show()
#bar Chart
df.plot(x="Month",y="sales",kind="bar",color="yellow")
plt.title("Monthly sales Charts")
plt.ylabel("sales")
plt.show()
#multiple line Charts
df.plot(x="Month",y=["sales","profits"],kind="line",marker=0)
plt.title("Sales vs profits trend")
plt.show()
#pile charts
df.set_index("Month")["sales"].plot(kind="pie", autopct="%1.1f%%")
plt.title("Sales Distribution")
plt.ylabel("")
plt.show()
#histrogram
df["sales"].plot(kind="hist")
plt.title("Sales Distribution")
plt.show()
#scatter plot
df.plot(kind="scatter",x="sales",y="profits")
plt.title("Sales vs profits")
plt.show()
#box plot
df[["sales","profits"]].plot(kind="box")
plt.title("Sales vs profits")
plt.show()
#Stacked Bar Chart
df.plot(x="Month",y=["sales","profits"],kind="bar",stacked=True)
plt.title("Sales vs profits")
plt.show()
```
